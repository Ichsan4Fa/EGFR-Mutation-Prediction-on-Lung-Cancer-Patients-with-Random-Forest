```python
def remove_outliers(df, column):
    for cols in column:
        Q1 = df[cols].quantile(0.25)
        Q3 = df[cols].quantile(0.75)
        IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    df = df[(df[cols] <= upper_bound) & (df[cols] >= lower_bound)]
    return df
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
le = preprocessing.LabelEncoder()
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False).set_output(transform = "pandas")

def one_hot_encoder(df, column_name):
  one_hot_encoded = encoder.fit_transform(df[column_name])
  one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(column_name))
  df_encoded = pd.concat([df, one_hot_df], axis=1)
  encoded_df = df_encoded.drop(column_name, axis=1)
  return encoded_df
```

# Modeling

```python
import numpy as np
import pandas as pd
from collections import Counter
import warnings
warnings.filterwarnings("ignore")

class Node:
  def __init__(self, feature=None, threshold=None, left=None, right=None,*,value=None):
    self.feature = feature
    self.threshold = threshold
    self.left = left
    self.right = right
    self.value = value

  def is_leaf_node(self):
   return self.value is not None
```

```python
class decision_Tree:
    def __init__(self, min_samples_split=5, max_depth=10, n_features=None):
      self.n_features = n_features
      self.root = None
      self.min_samples_split = min_samples_split
      self.max_depth = max_depth

    def train(self, Data, kelas):
      self.n_features = Data.shape[1] if not self.n_features else min(self.n_features, Data.shape[1])
      self.root = self.grow_tree(Data, kelas)

    def predict(self,X):
      return np.array([self.Traverse(x,self.root) for x in X])

    def grow_tree(self, Data, kelas, depth=0):
      try:
        n_samples, n_feats = Data.shape
        n_labels = len(np.unique(kelas))
        if (depth >= self.max_depth or n_labels == 1 or n_samples < self.min_samples_split):
          leaf_value = self.most_common_label(kelas)
          return Node(value=leaf_value)

        feat_indexes = np.random.choice(n_feats, self.n_features, replace=False)
        best_feat, best_thresh = self.best_split(Data, kelas, feat_indexes)
        left_idxs, right_idxs = self.split(Data[:, best_feat], best_thresh)
        left = self.grow_tree(Data[left_idxs, :], kelas[left_idxs], depth+1)
        right = self.grow_tree(Data[right_idxs, :], kelas[right_idxs], depth+1)

        return Node(best_feat, best_thresh, left, right)
      except Exception as e:
        print(f"Error during tree growth: {e}")
        return None

    def best_split(self,X,y,feat_idxs):
        try:
            best_gain=-1
            split_idxs,split_thres=None,None
            for feat_idx in feat_idxs:
                X_column=X[:,feat_idx]
                thresholds=np.unique(X_column)
                for threshold in thresholds:
                    gain=self.information_gain(X_column,y,threshold)

                    if gain>best_gain:
                        best_gain=gain
                        split_idxs=feat_idx
                        split_thres=threshold
            return split_idxs,split_thres
        except Exception as e:
            print("error",e)

    def split(self, x_column, threshold):
      left_idx = np.argwhere(x_column <= threshold).flatten()
      right_idx = np.argwhere(x_column > threshold).flatten()
      return left_idx, right_idx

    def information_gain(self,X_column,y,threshold):
        try:
            parent_entropy=entropy(y)
            left_idxs,right_idxs=self.split(X_column,threshold)
            #handling empty values
            if len(left_idxs)==0 or len(right_idxs)==0:
                return 0
            n=len(y)
            n_l,n_r=len(left_idxs),len(right_idxs)
            e_l,e_r=entropy(y[left_idxs]),entropy(y[right_idxs])
            child_entropy=(n_l/n)*e_l +(n_r/n)*e_r
            return parent_entropy-child_entropy
        except Exception as e:
            print("error",e)

    def Traverse(self, x, node):
      if node is None:
           print("Node is None")
           return None
      if node.is_leaf_node():
        return node.value
      if x[node.feature] <= node.threshold:
        return self.Traverse(x, node.left)
      return self.Traverse(x, node.right)

    @staticmethod
    def most_common_label(y):
        counter = Counter(y)
        most_common = counter.most_common(1)[0][0]
        return most_common


def entropy(y):
  hist=np.bincount(y)
  ps=hist/len(y)
  entropy = -(np.sum(p *np.log2(p) for p in ps if p>0))
  return entropy
```

```python
class random_Forest_Classifier:
    def __init__(self, n_trees=150, min_samples_split=2, max_depth=50, n_features=None):
        self.n_trees = n_trees
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
        self.n_features = n_features
        self.trees = []

    def fit(self, X, y):
        self.trees = []
        for _ in range(self.n_trees):
            tree = decision_Tree(
                min_samples_split=self.min_samples_split, max_depth=self.max_depth, n_features=self.n_features
            )
            X_samp, y_samp = bootstrap_sample(X, y)
            tree.train(X_samp, y_samp)

            self.trees.append(tree)

    def predict(self, X):
      tree_preds = np.array([tree.predict(X) for tree in self.trees])
      tree_preds = np.swapaxes(tree_preds, 0, 1)
      y_pred = [most_common_label(tree_pred) for tree_pred in tree_preds]
      return np.array(y_pred)

    def predict_proba(self, X):
      y_predictions = np.array([tree.predict(X) for tree in self.trees])
      y_predictions = np.swapaxes(y_predictions, 0, 1)

      unique_cls = sorted(set(y for tree in self.trees for y in tree.predict(X)))

      probs = np.zeros((y_predictions.shape[0], len(unique_cls)))

      for i, sample_preds in enumerate(y_predictions):
        n_class = Counter(sample_preds)
        total_class = sum(n_class.values())
        for cls in unique_cls:
          probs[i, cls] = n_class.get(cls,0)/total_class
      return np.array(probs)

    def score(self,X,y_true):
      y_pred = self.predict(X)
      return np.mean(y_pred==y_true)

    def measure(self,X, y_true):
      y_pred = self.predict(X)
      tp = np.sum((y_true == 1) & (y_pred == 1))
      fp = np.sum((y_true == 0) & (y_pred == 1))
      fn = np.sum((y_true == 1) & (y_pred == 0))
      if tp + fp == 0 or tp + fn == 0:
        return 0.0
      else:
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        return 2 * (precision * recall) / (precision + recall)

    def set_params(self, **params):
      for param, value in params.items():
          setattr(self, param, value)
      return self

    def get_params(self, deep=True):
      params = {
          'n_trees': self.n_trees,
          'min_samples_split': self.min_samples_split,
          'max_depth': self.max_depth,
          'n_features': self.n_features
      }
      return params

def bootstrap_sample(X, y):
   n_samples = X.shape[0]
   idxs = np.random.choice(n_samples, n_samples, replace=True)
   return X[idxs], y[idxs]

def most_common_label(y):
   counter = Counter(y)
   most_common = counter.most_common(1)[0][0]
   return most_common
```
